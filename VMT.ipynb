```python
vmt_list = []
for path in vmt_paths:
    route_id = path[len("/Users/ishigurofutoshi/lab_materials/data/VMT/"):-5]
    tmp_df = pd.read_excel(path)
    tmp_df["length"] = length_df.loc["length",int(route_id)]
    tmp_df["Minimum/length"] = tmp_df.loc[:,"Minimum"] / tmp_df.loc[:,"length"]
    df = pd.concat([tmp_df],keys=[route_id],axis=1)
    vmt_list.append(df)
df = pd.concat(vmt_list,axis=1)
```

## 各ルートの最小のVMTを確認

```python
min_vmts = []
for route_id in [a for a,b in df.columns.values if b=="Minimum"]:
    min_vmts.append(df.loc[:,(route_id,"Minimum/length")].min())
#print(min_vmts)
min_vmts_df = pd.DataFrame(np.array(min_vmts).reshape(1,278),columns=list(set([a for a,b in df.columns.values])),index=["min_vmt"])
```
